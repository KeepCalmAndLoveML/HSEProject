### По типу фигуры 

**Песочные часы**
* Характеристики
    * Плечи и бёдра одинаковой ширины
    * Талия уже бедер хотя бы на 25 см
* Подходит
    * Подчёркивать свою талию: платье-футляр, юбка-карандаш, блузка
* Не подходит
    * Стиль "оверсайз": балахоны
    
    
    
**Треугольник (груша)**
* Характеристики
    * Небольшие объёмы в области груди
    * Узкие плечи
    * Широкие и женственные бёдра
    * Узкая талия
    * Разница между талией и бёдрами ~30 см
* Подходит
    * Добавить объёма в верхней части: блузки, рубашки, футболки, жакеты
    * V-образный вырез, жакет, или топ с баской: скрыть бедра и полные ягодицы
    * Юбка годе, А-образная, плотная, прямая юбка-карандаш
    * Прямые брюки, слегка зауженные брюки или джинсы клеш
    * Рекомендуются украшения: платки, броши в области шеи и груди
    * Обувь с устойчивым каблуком 
    * Наилучшим типом одежды будет раздельная одежда - юбка и блузка
    * Брюки должеы плотно сидеть на бёдрах и слегка расширяться к низу
* Не подходит
    * Не следует привлекать внимания к нижней части тела 
    * Облегающие платья 

**Перевернутый треугольник**
* Характеристики
    * Широкие плечи
    * Большая грудь
    * Узкие бёдра
* Подходит
    * Пышные юбки (А-образная, юбка-тюльпан, юбка-колокол)
    * Широкие брюки
    * Брючный костюм
* Не подходит 
    * Двубортные жакеты 

**Прямоугольник**
* Характеристики
    * Плечи и бёдра одинаковой ширины
    * Нет выраженной талии
* Подходит
    * V-образные вырезы
    * Плотная или прямая A-образная юбка
    * Прямые брюки
    * Зауженные брюки
    * Брючный костюм, однобортный жакет
* Не подходит
    * ?

**Круг (Яблоко)**
* Характеристики
    * Выступающий живот
    * Пышный бюст 
    * Полные плечи, руки, ноги
    * Изящные ноги ниже колена
* Подходит
    * V-образный вырез
    * Прямые юбки до колена и чуть ниже
    * Плотные брюки
    * Жакет-Кимоно
    * Однотонная туника со слегка зауженными брюками
* Не подходит
    * Обтягивающая одежда
    * Короткие куртки и пальто 
    * Широкие или длинные юбки

### Другие параметры

**Низкий рост**
* Подходит    
    * Юбка и блузка одного цвета 
    * Обувь на устойчивом каблуке 
    * Распхнутый жилет, кардиган
    * Брюки в пол
* Не подходит
    * Ботфорты и полусапожки
    * Длинные жакеты и кардиганы
    * Длинная юбка
    
**Высокий рост**
* Подходит
    * Удлиненный жакет
    * Укороченные брюки
    * Подвернутые брюки
    * Юбки миди

**Полнота**
* Подходит
    * V-образный вырез 
* Не подходит
    * Всё балахонистое и многослойное
    * Стиль оверсайз

**Худоба**
* Подходит
    * А-образный силуэт
    * Стиль оверсайз
